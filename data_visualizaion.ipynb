### Libraries and Dataset

```python
import pandas as pd
import seaborn as sns

sns.__version__
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
display(tips.head())
```

#### Play with visualization

##### Tips Data Set

```python
### relational plot

plot = sns.relplot(
    
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
    hue='sex'
    )
```

```python
### relational plot

plot = sns.relplot(
    
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
    hue='sex',
    row = 'day',
    col = 'sex',
    height = 4
    )
```

##### MRI Data Set

```python
mri = sns.load_dataset("fmri")
display(mri.head())
```

```python
mri_plot = sns.relplot(
    
    data=mri,
    x="timepoint",
    y='signal',
    kind= 'line',
    height= 6,
    # col= "region",
    hue='region'
)

mri_plot.set(
    
    xlabel="Time Plot",
    ylabel="Signal"
)
```
