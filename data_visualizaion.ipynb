```python
import pandas as pd
import seaborn as sns

sns.__version__
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
display(tips.head())
```

#### Play with visualization

##### Tips Data Set

```python
### relational plot

plot = sns.relplot(
    
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
    hue='sex'
    )
```

```python
### relational plot

plot = sns.relplot(
    
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
    hue='sex',
    row = 'day',
    col = 'sex',
    height = 4
    )
```
