```python
df = pd.read_csv("homelessness.csv")
display(df.head())
```

```python
### Sort data with one variable

sort_by = df.sort_values(by=["region"], ascending=[False])
display(sort_by.head(10))



### Sort data with multiple variable

sort_by = df.sort_values(by=["region", "state"], ascending=[False, False])
display(sort_by.head(10))
```
