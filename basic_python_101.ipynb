```python
##### Print the square of only the even numbers 

print("Square of Even numbers in the list are:")
for x in numbers:
    if x%2 ==0:
        print(x**2, end=" ")


print("\n\n")
###### Create a list using even numbers:

list_of_evens=[]
list_of_odds=[]
list_of_even_square=[]

for x in numbers:
    if x%2 ==0:
        print(x, end=" ")
        y=x**2
        list_of_even_square.append(y)
        list_of_evens.append(x)
    else:
        list_of_odds.append(x)
        
print("\nlist of even is:",list_of_evens)
print("\nlist of odds is:",list_of_odds)
print("\nlist of even square is:",list_of_even_square)
```

```python
all_numbers = [x for x in numbers] 
print("All Numbers:", all_numbers) 

evens = [x for x in numbers if x%2 ==0]
print("Even Numbers:", evens) 

square_of_all = [x**2 for x in numbers ]
print("Square of all Numbers:", square_of_all) 


square_of_evens = [x**2 for x in numbers if x %2 ==0]
print("Square of all Evens:", square_of_evens) 
```

```python
from tqdm import tqdm
import time
 
# Progress bar 1: Default settings
for i in tqdm(range(300)):
	time.sleep(0.03)
 
# Progress bar 2: Customized bar format and color
for i in tqdm(range(300), bar_format='[{elapsed}<{remaining}] {n_fmt}/{total_fmt} | {l_bar}{bar} {rate_fmt}{postfix}', colour='yellow'):
	time.sleep(0.03)
 
# Progress bar 3: Customized bar format and color, leave=False
for i in tqdm(range(300), bar_format='[{elapsed}<{remaining}] {n_fmt}/{total_fmt} | {l_bar}{bar} {rate_fmt}{postfix}', colour='red', leave=False):
	time.sleep(0.01)
```

#### Numpy Array

```python
########  With Zero
print("1D Array")
zero_np_array_1d = np.zeros(
    shape=(6,),
    dtype=np.int32,
    )

print(zero_np_array_1d)

print("\n2D Array")
zero_np_array_2d = np.zeros(
    shape=(5,2),
    dtype=np.int32,
    )

print(zero_np_array_2d)

print("\n3D Array")
zero_np_array_3d = np.zeros(
    shape=(5,2,3),
    dtype=np.int32,
    )

print(zero_np_array_3d)

```
